```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
```

```python
(X_train,y_train),(X_test,y_test)=keras.datasets.mnist.load_data()#load the digit detaset
```

Scaling the values to incerase the accuracy of the model

```python
X_train=X_train/255
X_test=X_test/255
```

```python
X_train_flattened=X_train.reshape(len(X_train),28*28)
X_test_flattened=X_test.reshape(len(X_test),28*28)
```

```python
X_train_flattened[0]
```

```python
model=keras.Sequential([
    keras.layers.Dense(10,input_shape=(784,),activation='sigmoid')#Dense means that one layer is connected to every other layer in the NN , 10 represents
    #the number of output layers where we kow that the number of input layers are 784
])
model.compile(optimizer="adam",#these help to train the model efficiently
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"]
              )
model.fit(X_train_flattened,y_train,epochs=5)
```

```python
model.evaluate(X_test_flattened,y_test)
```

```python
plt.matshow(X_test[0])
```

```python
y_predicted=model.predict(X_test_flattened)
y_predicted[0]
```

```python
np.argmax(y_predicted[0])#this is the predicted value
```

```python
y_predicted_labels=[np.argmax(i) for i in y_predicted]
y_predicted_labels[:5]
```

```python
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)
cm
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
model = keras.Sequential([
    keras.layers.Dense(100, input_shape=(784,), activation='relu'),  # Hidden layer
    keras.layers.Dense(10, activation='sigmoid')  # Output layer
])

# Compile the model
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

# Train the model
model.fit(X_train_flattened, y_train, epochs=5)
```

```python
model.evaluate(X_test_flattened,y_test)
```

```python
y_predicted=model.predict(X_test_flattened)
y_predicted_labels=[np.argmax(i) for i in y_predicted]
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),   # Flattens 28x28 → 784
    keras.layers.Dense(100, activation='relu'),  # Hidden layer
    keras.layers.Dense(10, activation='sigmoid') # Output layer (better than sigmoid)
])

model.compile(optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])

model.fit(X_train, y_train, epochs=5)  # use original X_train (28x28)
```
